```python
#importing libaries
import joblib
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
```

```python
#importing dataset
data = pd.read_csv('project4_waste_generation.csv')
```

```python
#Displaying first five rows
data.head()
```

```python
#Analyzing the dataset
data.info()
```

```python
#viewing data features
data.describe()
```

```python
#Checking for Null Values
data.isnull().sum()
```

```python
#Spliting data to remove target column values for efficient training 
features = data.drop((['Weekly_Waste_Weight_kg', 'Week_Start_Date', 'Neighborhood']), axis=1)
target = data['Weekly_Waste_Weight_kg']
```

```python
#Separating our data into training and testing datasets
features_train, features_test, target_train, target_test = train_test_split(features,target, test_size= 0.2, random_state=42)
```

```python
#training our model
model = LinearRegression()
model.fit (features_train, target_train)
```

```python
model_prediction = model.predict(features_test)
# features_test contains inputs the model has NEVER seen before
# model.predict() tells the model: "guess the waste volume for each row"
# model_predict is now a list of the model's guesses
```

```python
# mean_squared_error  → calculates MSE (average of squared errors)
# mean_absolute_error → calculates MAE (average of raw errors)
# r2_score            → calculates R² (how well model learned the pattern)

mae  = mean_absolute_error(target_test, model_prediction)
# MAE: compares y_test (real answers) vs y_predict (model guesses)
# Result = average kg the model is wrong by

mse  = mean_squared_error(target_test, model_prediction)
# MSE: same as MAE but SQUARES the errors before averaging
# This punishes big mistakes more heavily than small ones

rmse = np.sqrt(mse)
# RMSE: we take the square root of MSE to bring it back to kg units
# Easier to understand than MSE because it's in the same unit as your target

r2   = r2_score(target_test, model_prediction)
# R²: tells you what % of the waste pattern your model actually learned
# R² = 1.0 means perfect, R² = 0.87 means 87% accurate

print("Model Evaluation Results")
print("-------------------------")
print(f"MAE  : {mae:.2f} kg")   # :.2f means show only 2 decimal places
print(f"MSE  : {mse:.2f}")
print(f"RMSE : {rmse:.2f} kg")
print(f"R²   : {r2:.2f}")
```

```python
#Visualization
# Define colors for each metric bar
colors  = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12']
# blue=MAE, red=MSE, green=RMSE, orange=R²

# Create a figure with 2 side-by-side charts (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
# figsize=(14,5) means the whole figure is 14 inches wide and 5 inches tall
# axes[0] = left chart, axes[1] = right chart

# Add a main title at the top of the whole figure
fig.suptitle('Model Evaluation Metrics', fontsize=16, fontweight='bold')

# CHART 1 (LEFT) — BAR CHART FOR MAE, MSE, RMSE

axes[0].bar(
    ['MAE', 'MSE', 'RMSE'],        # labels on the x-axis
    [mae, mse, rmse],              # heights of the bars (the actual values)
    color=['#3498db', '#e74c3c', '#2ecc71'],  # colors for each bar
    edgecolor='black',             # black border around each bar
    width=0.4                      # how wide each bar is
)
axes[0].set_title('Error Metrics (Lower is Better)', fontsize=13)
# title for this specific chart

axes[0].set_ylabel('Value (kg)')
# label for the vertical axis

# Add the number on top of each bar so you can read the value clearly
for index, value in enumerate([mae, mse, rmse]):
    axes[0].text(index, value + 0.5, f"{value:.2f}", ha='center', fontweight='bold')
    # value + 0.5 means place the text slightly ABOVE the bar
    # ha='center' means center the text horizontally over the bar
    
# CHART 2 (RIGHT) — HORIZONTAL GAUGE BAR FOR R² SCORE

# Draw a light grey background bar from 0 to 1 (full range)
axes[1].barh(['R² Score'], [1], color='#ecf0f1', edgecolor='black', height=0.3, zorder=0)
# zorder=0 means draw this BEHIND everything else (it's just the background)
# Draw the actual R² score bar on top of the grey background
axes[1].barh(['R² Score'], [r2], color='#f39c12', edgecolor='black', height=0.3)
# This orange bar will be as long as your R² value (e.g. 0.87 = 87% of the way)

axes[1].set_xlim(0, 1)
# Force the x-axis to go from 0 to 1 (since R² is always between 0 and 1)

axes[1].set_title('R² Score (Higher is Better)', fontsize=13)

# Draw a green dashed vertical line at 0.8 to show the "good model" threshold
axes[1].axvline(x=0.8, color='green', linestyle='--', linewidth=1.5, label='Good threshold (0.8)')
# If your orange bar passes this green line, your model is good ✅

# Show the actual R² number next to the orange bar
axes[1].text(r2 + 0.02, 0, f"{r2:.2f}", va='center', fontsize=13, fontweight='bold', color='#f39c12')
# r2 + 0.02 places the text just to the RIGHT of where the bar ends

axes[1].legend()
# Show the legend explaining what the green dashed line means

# Automatically adjust spacing between the two charts so nothing overlaps
plt.tight_layout()

# Save the chart as a PNG image file in your project folder
plt.savefig('model_evaluation.png', dpi=150, bbox_inches='tight')
# dpi=150 means good quality image (higher dpi = sharper image)
# bbox_inches='tight' removes unnecessary white space around the chart

plt.show()
# Display the chart on screen

print("Graph saved as model_evaluation.png ✅")
```

```python
# CHART 3 — ACTUAL vs PREDICTED LINE CHART (separate figure)
plt.figure(figsize=(10, 5))
# Create a brand new figure (separate from the two charts above)
# 10 inches wide, 5 inches tall

# Draw the REAL waste values as a solid blue line
plt.plot(target_test.values, label='Actual Waste (kg)', color='#3498db', linewidth=2)
# y_test.values = the real correct answers from your test data
# label = what shows up in the legend

# Draw the MODEL's PREDICTED values as a dashed red line
plt.plot(model_prediction, label='Predicted Waste (kg)', color='#e74c3c', linewidth=2, linestyle='--')
# y_predict = the model's guesses
# linestyle='--' makes it a dashed line so you can tell them apart visually

plt.title('Actual vs Predicted Waste Volume', fontsize=14, fontweight='bold')
plt.xlabel('Sample Index')   # x-axis = each test sample (row number)
plt.ylabel('Waste Volume (kg)')  # y-axis = the waste volume in kg
plt.legend()  # show the legend (blue=actual, red=predicted)

plt.tight_layout()  # clean up spacing

# Save this chart too as a PNG for use in your PowerPoint
plt.savefig('actual_vs_predicted.png', dpi=150, bbox_inches='tight')

plt.show()  # display on screen

print("Graph saved as actual_vs_predicted.png ✅")
# These two PNG files will appear in your project folder automatically
# Use them in your PowerPoint on Slide 8 and Slide 9 !
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score  # ← correct imports for regression

# ── Train models ──────────────────────────────────────────
tree = DecisionTreeRegressor(random_state=42)
tree.fit(features_train, target_train)

forest = RandomForestRegressor(n_estimators=100, random_state=42)
forest.fit(features_train, target_train)

# ── Predictions ───────────────────────────────────────────
y_pred_tree   = tree.predict(features_test)
y_pred_forest = forest.predict(features_test)

# ── Metrics ───────────────────────────────────────────────
r2_tree    = r2_score(target_test, y_pred_tree)   * 100
r2_forest  = r2_score(target_test, y_pred_forest) * 100

mae_tree   = mean_absolute_error(target_test, y_pred_tree)
mae_forest = mean_absolute_error(target_test, y_pred_forest)

# ── Plot ──────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Model Results', fontsize=16, fontweight='bold')

# Chart 1 — R² Score Bar Chart
axes[0].bar(['Decision Tree', 'Random Forest'],
            [r2_tree, r2_forest],
            color=['#1a73e8', '#2ecc71'],
            edgecolor='white', width=0.4)
axes[0].set_title('R² Score (Higher is Better)')
axes[0].set_ylim(0, 110)
for i, v in enumerate([r2_tree, r2_forest]):
    axes[0].text(i, v + 1, f"{v:.1f}%", ha='center', fontweight='bold')

# Chart 2 — MAE Bar Chart
axes[1].bar(['Decision Tree', 'Random Forest'],
            [mae_tree, mae_forest],
            color=['#00bcd4', '#e91e8c'],
            edgecolor='white', width=0.4)
axes[1].set_title('MAE in kg (Lower is Better)')
for i, v in enumerate([mae_tree, mae_forest]):
    axes[1].text(i, v + 0.5, f"{v:.2f} kg", ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('results.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"Decision Tree  → R²: {r2_tree:.2f}%  | MAE: {mae_tree:.2f} kg")
print(f"Random Forest  → R²: {r2_forest:.2f}% | MAE: {mae_forest:.2f} kg")
```
